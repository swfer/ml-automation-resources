# Análisis de los datos utilizados como ejemplo

El presente cuaderno de Jupyter tiene como objetivo mostrar el análisis de la distribución de valores de la serie temporal utilizada como ejemplo en el proyecto principal y en otros cuadernos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
import warnings
warnings.filterwarnings('ignore')
```

Los datos a utilizar corresponden a un registro diario de la temperatura mínima en Melbourne (Australia) entre los años 1981 y 1990, cuya carga del csv proporcionado se realiza a continuación.

```python
data = pd.read_csv('data/daily-min-temperatures.csv', index_col=0)
```

```python
data
```

Una primera visualización de la serie completa se ofrece a través del siguiente gráfico temporal donde se observa una clara estacionalidad de los datos para cada año.

```python
data.plot(title="Temperaturas diarias mínimas (Cª) entre 1981 y 1990 en Melbourne (Australia)", figsize=(10,5))
```

Aplicando a un año determinado la visualización anterior, en concreto 1981, se puede observar como la variación atiende a las diferentes estaciones: descenso de la temperatura a partir del segundo trimestre (otoño), obteniendo los mínimos en los meses de junio y julio (invierno), y un progresivo aumento que se extiende desde agosto al primer trimestre del año siguiente (primavera y verano).

```python
data.iloc[0:364].plot(title="Temperaturas diarias mínimas (Cª) de 1981 en Melbourne (Australia)", figsize=(10,5))
```

Por otro lado, el diagrama de caja mostrado en la siguiente figura ofrece una visualización sobre cómo se distribuyen los datos entre los límites inferior y superior de 0 a 22.5 grados Celsius respectivamente, con valores anómalos que alcanzan los 26.3 y una mitad de las temperaturas registrada entre los 8 y 14 grados Celsius.

```python
data.boxplot(figsize=(10,5))
```

Esta información puede verse expresada también en las siguientes tablas que, además de ofrecer los datos anteriores mediante estadísticos, ayudan a comprender el instante concreto donde ocurren los extremos.
